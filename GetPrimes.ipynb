## やりたいこと
+ 1000以下の素数を列挙する

```python
%%sql
drop view if exists series;
create view series as (
    select generate_series(1,1000) as num
);
```

```python
%%sql
-- 整数の中で割り切れるようなものが存在しないものを列挙する
select *
from series as s1
where not exists (
    select *
    from series as s2
    where s2.num >= 2 and
    s1.num > s2.num and
    mod(s1.num, s2.num) = 0
)
limit 20
```

```python
%%sql
-- すべての素数から非素数を除外する
select *
from Series
except (
    select (s1.num * s2.num) as non_prime
    from Series as s1, Series as s2
    where s1.num between 2 and 500 and
    s2.num between 2 and 500 and
    s1.num <= s2.num
)
order by num
```
